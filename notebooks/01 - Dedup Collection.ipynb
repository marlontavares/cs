```
%sql
create or replace table aux_contrato_random as
select 
    a.*,
    b.I
from (
  select
    a.contrato,
    b.anomes,
    a.qtd,
    a.rnd
  from
    contrato_consolidado as a
    left join 
    (select anomes, NTILE(21) OVER (ORDER BY anomes) as I from (select distinct anomes from (select distinct contrato, anomes from base_estudo_collection_202007_202110) order by anomes)) as b) as a
left join
    (select distinct contrato, anomes, row_number() OVER(PARTITION BY contrato ORDER BY anomes) AS I from (select distinct contrato, anomes from base_estudo_collection_202007_202110) order by contrato, anomes) as b
    on a.contrato = b.contrato and a.anomes = b.anomes
```

```
%sql
create or replace temp view base_estudo_coll_dedup as 
select
  a.*,
  case when b.contrato is null then 1 else 0 end as fl_excl_dedup
from
  base_estudo_collection_202007_202110 as a
  left join base_dedup_collection as b
on a.contrato = b.contrato and a.anomes = b.anomes
-- retirando contrato duplicados por safra ou das safras de OOT e REC
where case when b.contrato is null then 1 else 0 end = 0 or a.anomes >= 202107
```

```
%sql
create or replace temp view dup_cntr_anomes as 
select contrato, anomes, count(*) as qtd from base_estudo_coll_dedup group by 1,2 having qtd = 2
```

```
%sql 
create or replace table base_estudo_collection_dedup as 
select
  a.*,
  case when b.contrato is null then 0 else 1 end as flg_dup_safra
from
  base_estudo_coll_dedup as a
left join (select contrato, anomes, last(data_posicao) as data_posicao from base_estudo_collection_dedup where contrato in (select contrato from dup_cntr_anomes) group by 1,2 order by 1) as b
on a.contrato = b.contrato and a.data_posicao = b.data_posicao
left join
  cntr_anomes_3mais as c
on a.contrato = c.contrato
where case when b.contrato is null then 0 else 1 end not in (1) and c.contrato is null
```
